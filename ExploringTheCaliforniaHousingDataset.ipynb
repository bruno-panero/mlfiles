```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##importo archivo
housing_data = pd.read_csv('datasets/housing.csv')
```

```python
##veo los primeros 5 registros y el encabezado
housing_data.head()
```

```python
##exporo la cantidad
housing_data.shape
```

```python
##borro los que tengan valores vacios
housing_data = housing_data.dropna()
```

```python
housing_data.shape
```

```python
##propiedades de los atributos numericos de la coleccion
housing_data.describe()
```

```python
##un distinct de los valores del campo señalado entre apostrofos
housing_data['ocean_proximity'].unique()
```

```python
##graficamos relacion entre la variable total de cuartos y el valor medio
fig, ax = plt.subplots(figsize=(12,8))
plt.scatter(housing_data['total_rooms'],housing_data['median_house_value'])
plt.xlabel('Total Rooms')
plt.ylabel('Median House Value')
```

```python
##graficamos relacion entre la variable antiguedad y el valor medio
fig, ax = plt.subplots(figsize=(12,8))
plt.scatter(housing_data['housing_median_age'],housing_data['median_house_value'])
plt.xlabel('Median Age')
plt.ylabel('Median House Value')
```

```python
##graficamos relacion entre la variable ingresos medios y el valor medio
fig, ax = plt.subplots(figsize=(12,8))
plt.scatter(housing_data['median_income']*10000,housing_data['median_house_value'])
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
```

```python
##aplicamos correlacion entre las variables del dataset
housing_data_corr = housing_data.corr()

##luego la visualizamos
housing_data_corr
```

```python
##graficamos la correlacion entre las variable
fig, ax = plt.subplots(figsize=(12,8))
sns.heatmap(housing_data_corr,annot=True)
```
